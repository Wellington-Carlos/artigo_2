## Fractais

Os fractais são padrões infinitamente complexos que são auto-similares em diferentes escalas. Eles são criados repetindo um processo simples repetidamente em um ciclo de feedback contínuo. Uma característica chave de um fractal é a sua dimensão fractal. Ao contrário da dimensão euclidiana, a dimensão fractal é geralmente expressa por um não inteiro e é um indicador da complexidade ou rugosidade de uma determinada figura. Alguns exemplos comuns incluem o **triângulo de Sierpiński com uma dimensão de Hausdorff de 1,585**, o **floco de neve de Koch com uma dimensão de Hausdorff de 1,262** e o **litoral da Grã-Bretanha, cuja dimensão fractal é de 1,21**.

```python
%run arquivos_py/Sierpinski_Triangle.py
```

Começamos com o triângulo de ordem 0 na figura acima. A próxima iteração, ordem 1, é composta por 3 triângulos menores. E a ordem 2 é composta por 9 triângulos. Portanto, cada iteração do fractal tem 3 vezes mais triângulos e N = 3. Em seguida, precisamos descobrir o fator de escala, r. Quanto menor é cada triângulo na ordem 1 do que na ordem 0? A aresta de cada triângulo na ordem 1 é metade do comprimento da aresta do triângulo na ordem 0. Então o fator de escala r=2. Sendo assim o triângulo de Sierpinski tem dimensão Hausdorff igual a:
$$d = \frac{log(N)}{log(r)} = \frac{log(3)}{log(2)} = 1,585$$

Se considerarmos cada passo da consrução da curva, notamos que para passar de uma linha para a seguinte, substituímos três segmentos por quatro de igual comprimento, ou seja, o comprimento total é multiplicado por 4/3. O limite da sucessão geométrica de razão 4/3 é o infinito, o que significa que a figura final (ou para que tende esta sucessão) terá um comprimento infinito (designado por Mandelbrot como "infinito interno").

Sabe-se que uma homotetia (ampliação ou redução de distâncias, áreas ou volumes a partir de um ponto fixo) de razão três multiplica os comprimentos por 3, as superfícies por $3^2 = 9$, e os volumes por $3^3 = 27$ (o que, generalizando, permite calcular o "volume" de um objecto de dimensão d por $3^d$). Ora, como temos $3^d = 4$ para a curva de Koch, temos:
$$d = \frac{log(4)}{log(3)} = 1,262$$

Se perguntarmos "Qual é o comprimento da costa da Grã-Bretanha", a resposta é que depende de quão de perto você olha para ela, ou do comprimento de sua vara de medição. Se você medir o litoral pegando um mapa e colocando uma régua ao redor da borda, poderá obter um certo valor para o perímetro. Mas se você andasse pelas praias da Grã-Bretanha e somasse todos os seus passos, chegaria a um perímetro muito diferente (e maior). Mas esse valor também é uma aproximação. Para obter um valor mais preciso, você teria que medir o comprimento ao redor de cada pedregulho, cada pedra, cada pedregulho e até mesmo cada grão de areia. E em um nível microscópico, a areia também é um fractal e não pode ser facilmente medida. A resposta é: o litoral fica cada vez mais longo à medida que você o mede mais de perto, e se aproxima do infinito. É por isso que a dimensão fractal é um conceito muito útil para descrever um litoral.

Para derivar a dimensão, usaremos um método semelhante ao dos fractais geométricos e veremos como o comprimento medido muda em função da escala. Para fazer isso, precisaremos medir o perímetro usando réguas de diferentes comprimentos. Para ser consistente, vamos definir r como o fator de ampliação, que é o inverso do comprimento da régua.

Quando usamos uma régua grande (r=1, um fator de ampliação pequeno), obtemos uma aproximação muito ruim, mostrada em roxo, e um valor para o perímetro de N=9. À medida que o comprimento da régua diminui, a ampliação r aumenta e o valor do perímetro N aumenta. Estamos interessados na taxa de variação do perímetro em função do comprimento da régua. Quanto mais sinuosa for a linha de costa, mais o perímetro aumentará à medida que a régua encolhe. Para entender melhor a relação, podemos representar graficamente o perímetro versus o fator de ampliação (ou o inverso do comprimento da régua). 

A dimensão fractal do litoral, d, é simplesmente a inclinação da linha no gráfico acima:

$$d = \frac{log(N)}{log(r)} = 1,21$$

## O conjunto de Mandelbrot e o mapa logístico

Um dos fractais mais interessantes, o **conjunto de Mandelbrot** é o conjunto de números complexos c para os quais a função $x_{n+1} = x^2_n + c$ não diverge para infinito quando iterada de $z=0$. O caminho de todas essas órbitas quando plotado nos dá uma imagem intuitiva e bastante fácil de entender. Todos os pontos dentro do **cardióide** principal possuem um único ponto fixo, todos os pontos dentro do **bulbo** principal possuem 2 pontos limite e, posteriormente, todos os outros bulbos representam o conjunto de números que possuem diferentes números de pontos limite. O mapa logístico é um mapeamento polinomial de grau 2, frequentemente citado como um exemplo arquetípico de como o comportamento caótico e complexo pode surgir de equações dinâmicas não lineares muito simples. Por simples manipulação algébrica, o mapa logístico $x_{n+1} = rx_n(1 - x_n)$, pode ser recodificado na forma $x_{n+1} = x^2_n + c$. Isso nos leva a explorar ainda mais uma relação que produz uma outra relação bastante interessante quando mudamos nosso ponto de vista. A partir da estrutura de ramificação do mapa de bifurcação logístico podemos ler o número do ciclo correspondente ao conjunto de Mandelbrot.

## Caos Determinístico

O caos determinístico é o estudo de como os sistemas que seguem leis simples, diretas e determinísticas podem exibir um comportamento de longo prazo muito complicado e aparentemente aleatório. Um dos fundamentos da teoria do caos é a descoberta de Edward Lorenz de sistemas onde a dinâmica é sensível a pequenas mudanças nas condições iniciais, portanto, são imprevisíveis. Devido a esta sensibilidade o comportamento dos sistemas parece ser aleatório embora o modelo seja determinístico, o que significa que está bem definido e não contém quaisquer parâmetros aleatórios. No início, essa teoria era usada apenas no campo da meteorologia, mas logo se percebeu que ela também pode definir outros sistemas caóticos em diversas disciplinas científicas que se baseavam na previsibilidade do comportamento futuro dos sistemas.

## A relação entre o Mapa Logístico e o Caos

A equação logística é da forma $x_{n+1} = r x_n (1 - x_n)$. Mudando os valores de $r$ observamos os seguintes comportamentos:

* Quando $r \in [0,1]$ a sequência convergirá para zero.
* Quando $r \in [1,3)$ a sequência se aproxima do valor $r-\frac{1}{r}$, independentemente do valor de $x_0$, embora a taxa de convergência da função diminua quando o valor de $r$ aumenta.
* Quando $r \in [3,1 + \sqrt{6})$ a função oscila entre 2 pontos limites cujos valores são determinados por $r$.
* Quando $r \in [1 + \sqrt{6}, \sim(3,54409)]$ a sequência terá aproximações oscilantes entre 4 pontos limites.
* Quando $r > \sim(3,54409)$ a função terá aproximações oscilantes entre 8 pontos limites, depois 16, 32, e assim por diante.

A razão entre o dois sucessivos intervalos de bifurcação é conhecida como **constante de Feigenbaum**: $\delta=\lim\limits_{x\to\infty}\frac{a_{n-1}-a_{n-2}}{a_{n}-a_{n-1}}=4,6692016...$.

* Em $r\approx 3,56995$ é o início do caos.

Não vemos mais oscilações de período finito. Pequenas variações no valor inicial de $x_0$ produzem resultados drasticamente diferentes ao longo do tempo. Para valores de $r$ além de $\sim 3,56995$ a sequência exibe comportamento caótico, mas existe ainda certos valores de $r$ que apresentam comportamento não-caótico, algumas vezes chamados de ilhas de estabilidade. Por exemplo, para $r=1+\sqrt{8}$ a sequência oscila entre 3 valores, e com valores ligeiramente mais altos de $r$ a sequência oscila entre 6 pontos limites, 12, etc.

## Ruído versus Caos determinístico

O ruído é a variação aleatória de valores enquanto o caos acontece quando as variáveis de estado iniciais do sistema
diferem muito ligeiramente, o que leva a diferenças drásticas dos resultados tornando impossível prever o valor inicial
da saída.

Também podemos diferenciar entre ruído e caos determinístico usando fractais. A dimensão de Hausdorff do atrator de um modelo aleatório é geralmente infinita. Mas o atrator de um modelo de caos determinístico sempre nos dá um valor não inteiro, finito. Portanto, se a dimensão de Hausdorff de um modelo é finita, é um bom indicador de que é um modelo determinístico que pode ser representado por um sistema de equações diferenciais não lineares.

## O Problema dos três corpos

O problema dos três corpos é um **sistema de equações diferenciais ordinárias** que modelam três corpos de massas sob gravitação mútua em duas ou três dimensões. O movimento de tal sistema foi primeiro ponderado por Newton e reconsiderado por muitos matemáticos e cientistas incluindo Euler, Lagrange, Jacobi e Poincaré. O famoso problema dos três corpos teve uma grande influência na **física**, **matemática** e **dinâmica não linear**. Ela abriu caminho para um novo campo na ciência moderna, a **dinâmica caótica**.

Na física e na mecânica clássica, o **problema de três corpos** é o problema de tomar as **posições e velocidades iniciais (ou momentos)** de três massas pontuais e resolver seu movimento subsequente de acordo com as **leis do movimento de Newton** e a **lei da gravitação universal** de Isaac Newton. O problema de três corpos é um caso especial do **Problema de n-corpos**. Ao contrário do problema de dois corpos, não existe uma solução geral de forma fechada, pois o sistema dinâmico resultante é **caótico para a maioria das condições iniciais**, e métodos numéricos são geralmente necessários.

Historicamente, o primeiro problema específico de três corpos a receber estudo prolongado foi o que envolvia a **Lua, Terra e o Sol**. Em um sentido moderno estendido, um problema de três corpos é qualquer problema em mecânica clássica ou mecânica quântica que modela o movimento de três partículas.

O **problema dos dois corpos foi analisado por Johannes Kepler em 1609 e resolvido por Isaac Newton em 1687**. O problema dos três corpos foi um tópico central na física matemática de meados dos anos 1700 até o início dos anos 1900. Vários resultados exatos foram obtidos - notadamente a existência de configurações estáveis de triângulos equiláteros correspondentes aos chamados pontos de Lagrange. Muitos cálculos práticos aproximados, particularmente no sistema Terra-Lua-Sol, foram feitos usando expansões em série envolvendo milhares de termos algébricos. (Agora é possível obter a maioria dos resultados apenas por computação numérica direta). A partir de sua configuração básica, o sistema de três corpos conserva quantidades mecânicas padrão, como energia e momento angular. Mas pensava-se que também poderia conservar outras quantidades (ou as chamadas integrais do movimento). Em 1887, no entanto, Heinrich Bruns mostrou que não poderia haver tais quantidades exprimíveis como funções algébricas das posições e velocidades dos corpos (em coordenadas cartesianas padrão). Em meados da década de 1890, Henri Poincaré mostrou que também não poderia haver tais quantidades analíticas em posições, velocidades e razões de massa. E a partir desses resultados chegou-se à conclusão de que o problema dos três corpos não poderia ser resolvido em termos de fórmulas algébricas e integrais. Em 1912 , Karl Sundman no entanto, encontrou uma série infinita que poderia, em princípio, ser somada para dar a solução - mas que converge excepcionalmente lentamente. 

## Soluções para casos especiais

* 1767 - Leonhard Euler encontrou 3 famílias de soluções periódicas (massas colineares).
* 1772 - Joseph-Louis Lagrange encontrou 1 família de solução (massas formam um triângulo equilátero).
* 1892-1899 - Henri Poincaré estabeleceu a existência de um número infinito de soluções períodicas.
* 1893 - Meissel estabeleceu o problema de três corpos de Pitágoras (massas na razão 3:4:5 colocadas em repouso nos vértices de um triângulo retângulo 3:4:5). Burrau (1913) também investigou o problema. Victor Szebehely e Frederick Peters (1967) usando integração numérica encontraram uma solução periódica.
* 1912 - Karl Sundman provou que existe uma solução analítica na forma de uma série de potências (tem pouca utilidade prática devido a convergência hiper lenta).
* 1970 - Michel Hénon e Roger Broucke encontraram um conjunto de soluções.
* 1993 - Cristopher Moore solução numérica para momento angular zero com as massas movendo-se em torno de uma forma de 8.
* 2000 - Alain Chenciner e Richard Montgomery provou a solução de Moore, porém com probabilidade de existência real de 1%.
* 2013 - Milovan Suvakov e Veljko Dmitrasinovic descobriram 13 novas famílias de soluções com momento angular zero.
* 2015 - Ana Hudomal descobriu 14 novas famílias de soluções com corpos de massas iguais e momento angular zero.
* 2017 - Xiaoming Li e Shijun Liao encontraram 669 novas órbitas períodicas (massas iguais).
* 2018 - Xiaoming Li e Shijun Liao encontraram 1223 soluções adicionais (massas desiguais).
* 2020 - Xiaoming Li, Shijun Liao e Xiaochen Li encontraram 13.315 órbitas períodicas (massas desiguais).
* 2022 - Xiaoming Li, Shijun Liao e Yu Yang: 135.445 novas órbitas períodicas (HPC + ANN)
