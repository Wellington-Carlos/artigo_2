## Fractais

Os fractais so padres infinitamente complexos que so auto-similares em diferentes escalas. Eles so criados repetindo um processo simples repetidamente em um ciclo de feedback contnuo. Uma caracterstica chave de um fractal  a sua dimenso fractal. Ao contrrio da dimenso euclidiana, a dimenso fractal  geralmente expressa por um no inteiro e  um indicador da complexidade ou rugosidade de uma determinada figura. Alguns exemplos comuns incluem o **tringulo de Sierpiski com uma dimenso de Hausdorff de 1,585**, o **floco de neve de Koch com uma dimenso de Hausdorff de 1,262** e o **litoral da Gr-Bretanha, cuja dimenso fractal  de 1,21**.

```python
%run arquivos_py/Sierpinski_Triangle.py
```

Comeamos com o tringulo de ordem 0 na figura acima. A prxima iterao, ordem 1,  composta por 3 tringulos menores. E a ordem 2  composta por 9 tringulos. Portanto, cada iterao do fractal tem 3 vezes mais tringulos e N = 3. Em seguida, precisamos descobrir o fator de escala, r. Quanto menor  cada tringulo na ordem 1 do que na ordem 0? A aresta de cada tringulo na ordem 1  metade do comprimento da aresta do tringulo na ordem 0. Ento o fator de escala r=2. Sendo assim o tringulo de Sierpinski tem dimenso Hausdorff igual a:
$$d = \frac{log(N)}{log(r)} = \frac{log(3)}{log(2)} = 1,585$$

O limite da sucesso geomtrica de razo 4/3  o infinito, o que significa que a figura final (ou para que tende esta sucesso) ter um comprimento infinito (designado por Mandelbrot como "infinito interno").

Sabe-se que uma homotetia (ampliao ou reduo de distncias, reas ou volumes a partir de um ponto fixo) de razo trs multiplica os comprimentos por 3, as superfcies por $3^2 = 9$, e os volumes por $3^3 = 27$ (o que, generalizando, permite calcular o "volume" de um objecto de dimenso d por $3^d$). Ora, como temos $3^d = 4$ para a curva de Koch, temos:
$$d = \frac{log(4)}{log(3)} = 1,262$$

Estamos interessados na taxa de variao do permetro em funo do comprimento da rgua. Quanto mais sinuosa for a linha de costa, mais o permetro aumentar  medida que a rgua encolhe. ## O conjunto de Mandelbrot e o mapa logstico

Um dos fractais mais interessantes, o **conjunto de Mandelbrot**  o conjunto de nmeros complexos c para os quais a funo $x_{n+1} = x^2_n + c$ no diverge para infinito quando iterada de $z=0$. O caminho de todas essas rbitas quando plotado nos d uma imagem intuitiva e bastante fcil de entender. Todos os pontos dentro do **cardiide** principal possuem um nico ponto fixo, todos os pontos dentro do **bulbo** principal possuem 2 pontos limite e, posteriormente, todos os outros bulbos representam o conjunto de nmeros que possuem diferentes nmeros de pontos limite. O mapa logstico  um mapeamento polinomial de grau 2, frequentemente citado como um exemplo arquetpico de como o comportamento catico e complexo pode surgir de equaes dinmicas no lineares muito simples. Por simples manipulao algbrica, o mapa logstico $x_{n+1} = rx_n(1 - x_n)$, pode ser recodificado na forma $x_{n+1} = x^2_n + c$. Isso nos leva a explorar ainda mais uma relao que produz uma outra relao bastante interessante quando mudamos nosso ponto de vista. A partir da estrutura de ramificao do mapa de bifurcao logstico podemos ler o nmero do ciclo correspondente ao conjunto de Mandelbrot.

## Caos Determinstico

O caos determinstico  o estudo de como os sistemas que seguem leis simples, diretas e determinsticas podem exibir um comportamento de longo prazo muito complicado e aparentemente aleatrio. Um dos fundamentos da teoria do caos  a descoberta de Edward Lorenz de sistemas onde a dinmica  sensvel a pequenas mudanas nas condies iniciais, portanto, so imprevisveis. Devido a esta sensibilidade o comportamento dos sistemas parece ser aleatrio embora o modelo seja determinstico, o que significa que est bem definido e no contm quaisquer parmetros aleatrios. No incio, essa teoria era usada apenas no campo da meteorologia, mas logo se percebeu que ela tambm pode definir outros sistemas caticos em diversas disciplinas cientficas que se baseavam na previsibilidade do comportamento futuro dos sistemas.

## A relao entre o Mapa Logstico e o Caos

A equao logstica  da forma $x_{n+1} = r x_n (1 - x_n)$. Mudando os valores de $r$ observamos os seguintes comportamentos:

* Quando $r \in [0,1]$ a sequncia convergir para zero.
* Quando $r \in [1,3)$ a sequncia se aproxima do valor $r-\frac{1}{r}$, independentemente do valor de $x_0$, embora a taxa de convergncia da funo diminua quando o valor de $r$ aumenta.
* Quando $r \in [3,1 + \sqrt{6})$ a funo oscila entre 2 pontos limites cujos valores so determinados por $r$.
* Quando $r \in [1 + \sqrt{6}, \sim(3,54409)]$ a sequncia ter aproximaes oscilantes entre 4 pontos limites.
* Quando $r > \sim(3,54409)$ a funo ter aproximaes oscilantes entre 8 pontos limites, depois 16, 32, e assim por diante.

A razo entre o dois sucessivos intervalos de bifurcao  conhecida como **constante de Feigenbaum**: $\delta=\lim\limits_{x\to\infty}\frac{a_{n-1}-a_{n-2}}{a_{n}-a_{n-1}}=4,6692016...$.

* Em $r\approx 3,56995$  o incio do caos.

No vemos mais oscilaes de perodo finito. Pequenas variaes no valor inicial de $x_0$ produzem resultados drasticamente diferentes ao longo do tempo. Para valores de $r$ alm de $\sim 3,56995$ a sequncia exibe comportamento catico, mas existe ainda certos valores de $r$ que apresentam comportamento no-catico, algumas vezes chamados de ilhas de estabilidade. Por exemplo, para $r=1+\sqrt{8}$ a sequncia oscila entre 3 valores, e com valores ligeiramente mais altos de $r$ a sequncia oscila entre 6 pontos limites, 12, etc.

## O Problema dos trs corpos

O problema dos trs corpos  um **sistema de equaes diferenciais ordinrias** que modelam trs corpos de massas sob gravitao mtua em duas ou trs dimenses. O movimento de tal sistema foi primeiro ponderado por Newton e reconsiderado por muitos matemticos e cientistas incluindo Euler, Lagrange, Jacobi e Poincar. O famoso problema dos trs corpos teve uma grande influncia na **fsica**, **matemtica** e **dinmica no linear**. Ela abriu caminho para um novo campo na cincia moderna, a **dinmica catica**.

Na fsica e na mecnica clssica, o **problema de trs corpos**  o problema de tomar as **posies e velocidades iniciais (ou momentos)** de trs massas pontuais e resolver seu movimento subsequente de acordo com as **leis do movimento de Newton** e a **lei da gravitao universal** de Isaac Newton. O problema de trs corpos  um caso especial do **Problema de n-corpos**. Ao contrrio do problema de dois corpos, no existe uma soluo geral de forma fechada, pois o sistema dinmico resultante  **catico para a maioria das condies iniciais**, e mtodos numricos so geralmente necessrios.

Historicamente, o primeiro problema especfico de trs corpos a receber estudo prolongado foi o que envolvia a **Lua, Terra e o Sol**. Em um sentido moderno estendido, um problema de trs corpos  qualquer problema em mecnica clssica ou mecnica quntica que modela o movimento de trs partculas.

O **problema dos dois corpos foi analisado por Johannes Kepler em 1609 e resolvido por Isaac Newton em 1687**. O problema dos trs corpos foi um tpico central na fsica matemtica de meados dos anos 1700 at o incio dos anos 1900. Vrios resultados exatos foram obtidos - notadamente a existncia de configuraes estveis de tringulos equilteros correspondentes aos chamados pontos de Lagrange. Muitos clculos prticos aproximados, particularmente no sistema Terra-Lua-Sol, foram feitos usando expanses em srie envolvendo milhares de termos algbricos. (Agora  possvel obter a maioria dos resultados apenas por computao numrica direta). A partir de sua configurao bsica, o sistema de trs corpos conserva quantidades mecnicas padro, como energia e momento angular. Mas pensava-se que tambm poderia conservar outras quantidades (ou as chamadas integrais do movimento). Em 1887, no entanto, Heinrich Bruns mostrou que no poderia haver tais quantidades exprimveis como funes algbricas das posies e velocidades dos corpos (em coordenadas cartesianas padro). Em meados da dcada de 1890, Henri Poincar mostrou que tambm no poderia haver tais quantidades analticas em posies, velocidades e razes de massa. E a partir desses resultados chegou-se  concluso de que o problema dos trs corpos no poderia ser resolvido em termos de frmulas algbricas e integrais. Em 1912 , Karl Sundman no entanto, encontrou uma srie infinita que poderia, em princpio, ser somada para dar a soluo - mas que converge excepcionalmente lentamente.

* 1912 - Karl Sundman provou que existe uma soluo analtica na forma de uma srie de potncias (tem pouca utilidade prtica devido a convergncia hiper lenta).

* 2013 - Milovan Suvakov e Veljko Dmitrasinovic descobriram 13 novas famlias de solues com momento angular zero.
* 2015 - Ana Hudomal descobriu 14 novas famlias de solues com corpos de massas iguais e momento angular zero.
